```python
def shuffled_deck():
        deck = []
        for suit in ['C', 'D', 'H', 'S']:
                for num in range(6, 15):
                        deck.append(  [(num), suit]  )
        random.shuffle(deck)
        return deck
```

```python
def card_name(card):
    if card == 11:
        return 'J'
    elif card == 12:
        return 'Q'
    elif card == 13:
        return 'K'
    elif card == 14:
        return 'A'
    else:
        return str(card)
```
